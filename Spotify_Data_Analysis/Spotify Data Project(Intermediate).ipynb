#### Group by artist and calculate:

#### number of tracks

```python
each_album_count=spotify_data.groupby(['artist_name','album_name'])['album_total_tracks'].max().reset_index()
each_album_count
```

```python
total_track_count=each_album_count.groupby('artist_name')['album_total_tracks'].sum().reset_index(name='total_tracks_across_all_albums')
total_track_count.sort_values(by='total_tracks_across_all_albums',ascending=False)
```

#### Group by artist and calculate:

#### average track popularity

```python
round(spotify_data.groupby('artist_name')['track_popularity'].mean().reset_index().sort_values(by='track_popularity', ascending=False),2)
```

```python
spotify_data.groupby('artist_name')['track_duration_min'].mean().sort_values(ascending=False)
```

```python
spotify_tracks=spotify_data[['track_id','track_name','track_popularity']]
spotify_tracks
```

```python
spotify_data['track_popularity'].plot(kind='hist',bins=20)
plt.xlabel('Track Popularity')
plt.show()
```

```python
spotify_data.plot(kind='scatter',x='track_duration_min',y='track_popularity')
plt.show()
```

```python
spotify_data.groupby('artist_name')['track_popularity'].mean().sort_values(ascending=False)\
.head(10).plot(kind='bar')
plt.ylim(80,100)     

plt.show()
```

Most of the volume of track popularity are clustered around 40-80 indicating high popularity scores. The artist followers histogram scale is in tens of millions and majority of artists are around 1-20 million range. The scatter plot indicates that most tracks are around 2-6 minutes and those tracks are usually doing pretty good popularity wise.
